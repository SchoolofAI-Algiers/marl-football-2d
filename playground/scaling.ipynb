# modules

```python
import numpy as np
import matplotlib.pyplot as plt
```

# scaling

## utils

```python
def plot_scale(f, exp=1):
    n_values = np.arange(2, 24, 2)
    scaled_values = f(n_values, exp)
    
    plt.figure(figsize=(6, 4))
    plt.plot(n_values, scaled_values, marker='o', linestyle='-')
    plt.xlabel('Number of Players (n)')
    plt.ylabel('Scaling Factor sqrt(n / 22)')
    plt.title('Scaling Function sqrt(n / 22)')
    plt.grid()
    plt.show()
```

## functions

```python
def log_scaling(n, exp):
    return (np.log1p(n) / np.log1p(22)) ** exp
```

## experiment

```python
plot_scale(log_scaling, exp=0.1)
```
